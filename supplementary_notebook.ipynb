In this notebook, the differences of the performance of the non-linear fwDTI fitting procedure when different optional inputs are given is accessed. Then we also compare the non-linear fwDTI fitting performance when the scipy.optimization.least_squares with boundaries is used to solve the fwDTI model instead of the scipy.optimization.leastsq.

First, the relevant modules are imported:

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import time
import sys
%matplotlib inline

# To sample the diffusion-weighted directions
from dipy.core.sphere import disperse_charges, Sphere, HemiSphere

# To reconstruct the tables with the acquisition information
from dipy.core.gradients import gradient_table

# To perform the monte-carlo simulations
from dipy.sims.voxel import multi_tensor

# To process the diffusion tensor
import dipy.reconst.dti as dti

# Importing procedures to fit the free water elimination DTI model
from functions import (wls_fit_tensor, nls_fit_tensor, nls_fit_tensor_bounds)
```

Lets run the simulations of the article's Figure 1 only with one sample of each 120 diffusion tensor directions to speed computing time.

```python
# ---------------------------------------------------------------
# acquisition parameters
# ---------------------------------------------------------------

SNR = 40

# Sample 32 diffusion-weighted directions
n_pts = 32
theta = np.pi * np.random.rand(n_pts)
phi = 2 * np.pi * np.random.rand(n_pts)
hsph_initial = HemiSphere(theta=theta, phi=phi)
hsph_updated, potential = disperse_charges(hsph_initial, 5000)
directions = hsph_updated.vertices # directions for each shell

# Create full dataset parameters
# (6 b-values = 0, 32 directions for each b-values 500 and 1500)
bvals = np.hstack((np.zeros(6), 500 * np.ones(n_pts), 1500 * np.ones(n_pts)))
bvecs = np.vstack((np.zeros((6, 3)), directions, directions))
gtab = gradient_table(bvals, bvecs)
design_matrix = dti.design_matrix(gtab)

# ---------------------------------------------------------------
# Ground truth values of tissue's diffusion tensor and
# water volume fraction
# ---------------------------------------------------------------

# free water component
VF = np.linspace(0, 100, num=11)
Dwater = 3e-3

# tissue component
FA = np.array([0.71, 0.30, 0.22, 0.11, 0.])
L1 = np.array([1.6e-3, 1.080e-3, 1.000e-3, 0.900e-3, 0.8e-03])
L2 = np.array([0.5e-3, 0.695e-3, 0.725e-3, 0.763e-3, 0.8e-03])
L3 = np.array([0.3e-3, 0.625e-3, 0.675e-3, 0.738e-3, 0.8e-03])

# simulations are repeated for 120 different diffusion tensor's
# directions (and each direction repeated 100 times)
nDTdirs = 120
theta = np.pi * np.random.rand(nDTdirs)
phi = 2 * np.pi * np.random.rand(nDTdirs)
hsph_initial = HemiSphere(theta=theta, phi=phi)
hsph_updated, potential = disperse_charges(hsph_initial, 5000)
DTdirs = hsph_updated.vertices # directions for each shell

nrep = 1  # number of repetitions for each direction

# ----------------------------------------------------------------
# Generate simulations
# ----------------------------------------------------------------

DWI_simulates = np.empty((FA.size, VF.size, nrep * nDTdirs,
                          bvals.size))

for fa_i in range(FA.size):
    mevals = np.array([[L1[fa_i], L2[fa_i], L3[fa_i]],
                       [Dwater, Dwater, Dwater]])
    for vf_i in range(VF.size):
        fractions = [100 - VF[vf_i], VF[vf_i]]
        for di in range(nDTdirs):
            d = DTdirs[di]
            for s_i in np.arange(di * nrep, (di+1) * nrep):
                signal, sticks = multi_tensor(gtab, mevals,
                                              S0=100,
                                              angles=[d, (1, 0, 0)],
                                              fractions=fractions,
                                              snr=SNR)
                DWI_simulates[fa_i, vf_i, s_i, :] = signal
```

Similar to what was done on the article, the first guess of the models parameters is computed using the WLS procedure. For a processing time reference, the time necessary to compute the WLS procedure is measured.

```python
# ---------------------------------------------------------------
# WLS procedures is used to estimate parameters initial guess
# ---------------------------------------------------------------

# find S0 from data
S0 = np.mean(DWI_simulates[:, :, :, bvals==0], axis=3)

t0 = time.time()
# WLS procedures is used to estimate parameters initial guess
fw_params_initial, S0e = wls_fit_tensor(design_matrix, DWI_simulates,
                                        Diso=Dwater, S0=S0)
tr = time.time() - t0
print(tr)
```

The results of the first initial guess are plotted below:

```python
# ---------------------------------------------------------------
# Process diffusion tensor standard measures and plot
# ---------------------------------------------------------------

# Compute standard measures of the tissue's diffusion tensor
evals = fw_params_initial[..., :3]
fa = dti.fractional_anisotropy(evals)
f = fw_params_initial[..., 12]

median_fa = np.empty(VF.size)
lower_p = np.empty(VF.size)
upper_p = np.empty(VF.size)

# Plot FA values
colors = {0: 'r', 1: 'magenta', 2: 'black', 3: 'b', 4: 'g'}
plt.figure(figsize=(5, 4))
plt.axis([-0.1, 1.2, -0.1, 1.2])
for fa_i in range(FA.size):
    for vf_i in range(VF.size):
        median_fa[vf_i] = np.median(fa[fa_i, vf_i, :])
        p25, p75 = np.percentile(fa[fa_i, vf_i, :], [25 ,75])
        lower_p[vf_i] = median_fa[vf_i] - p25 
        lower_p[vf_i] = p75 - median_fa[vf_i]
    plt.errorbar(VF/100, median_fa, fmt='.',
                 yerr=[lower_p, lower_p],
                 color=colors[fa_i],
                 ecolor=colors[fa_i],
                 linewidth=1.0,
                 label='$FA: %.2f$' % FA[fa_i])
plt.xlabel('Simulated f-value')
plt.ylabel('Estimated FA')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

# Plot F values
median_f = np.empty(VF.size)
lower_p = np.empty(VF.size)
upper_p = np.empty(VF.size)

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

for idx, fa_i in enumerate([0, 4]):
    for vf_i in range(VF.size):
        median_f[vf_i] = np.median(f[fa_i, vf_i, :])
        p25, p75 = np.percentile(f[fa_i, vf_i, :], [25 ,75])
        lower_p[vf_i] = median_f[vf_i] - p25 
        lower_p[vf_i] = p75 - median_f[vf_i]
    axs[idx].errorbar(VF/100, median_f, fmt='.',
                      yerr=[lower_p, lower_p],
                      color=colors[fa_i],
                      ecolor=colors[fa_i],
                      linewidth=3.0,
                      label='$FA: %.2f$' % FA[fa_i])
    # plot ground truth lines
    axs[idx].plot([0, 1], [0, 1], 'b', label = 'Simulated f-value')
    axs[idx].legend(loc='upper left')
    axs[idx].set_ylim([-0.1, 1.2])
    axs[idx].set_xlim([-0.1, 1.2])
    axs[idx].set_xlabel('Simulated f-value')
    axs[idx].set_ylabel('Estimated f-value')
```

Now lets process and plot the results for the NLS procedure without imposing constraints in the diffusion tensor and without the use of the analytical jacobian function (funtions defaults):

```python
fw_params = fw_params_initial.copy()

t0 = time.time()
# NLLS procedures is used to find final parameters guess
fw_params, S0e = nls_fit_tensor(design_matrix, DWI_simulates, 
                                fw_params=fw_params, S0=S0,
                                Diso=Dwater, cholesky=False,
                                f_transform=True, jac=False,
                                mdreg=2.7e-3)

tf = time.time() - t0
print(tf / tr)

# Compute standard measures of the tissue's diffusion tensor
evals = fw_params[..., :3]
fa = dti.fractional_anisotropy(evals)
f = fw_params[..., 12]

median_fa = np.empty(VF.size)
lower_p = np.empty(VF.size)
upper_p = np.empty(VF.size)

# Plot FA values
colors = {0: 'r', 1: 'magenta', 2: 'black', 3: 'b', 4: 'g'}
plt.figure(figsize=(5, 4))
plt.axis([-0.1, 1.2, -0.1, 1.2])
for fa_i in range(FA.size):
    for vf_i in range(VF.size):
        median_fa[vf_i] = np.median(fa[fa_i, vf_i, :])
        p25, p75 = np.percentile(fa[fa_i, vf_i, :], [25 ,75])
        lower_p[vf_i] = median_fa[vf_i] - p25 
        lower_p[vf_i] = p75 - median_fa[vf_i]
    plt.errorbar(VF/100, median_fa, fmt='.',
                 yerr=[lower_p, lower_p],
                 color=colors[fa_i],
                 ecolor=colors[fa_i],
                 linewidth=1.0,
                 label='$FA: %.2f$' % FA[fa_i])
plt.xlabel('Simulated f-value')
plt.ylabel('Estimated FA')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

# Plot F values
median_f = np.empty(VF.size)
lower_p = np.empty(VF.size)
upper_p = np.empty(VF.size)

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

for idx, fa_i in enumerate([0, 4]):
    for vf_i in range(VF.size):
        median_f[vf_i] = np.median(f[fa_i, vf_i, :])
        p25, p75 = np.percentile(f[fa_i, vf_i, :], [25 ,75])
        lower_p[vf_i] = median_f[vf_i] - p25 
        lower_p[vf_i] = p75 - median_f[vf_i]
    axs[idx].errorbar(VF/100, median_f, fmt='.',
                      yerr=[lower_p, lower_p],
                      color=colors[fa_i],
                      ecolor=colors[fa_i],
                      linewidth=3.0,
                      label='$FA: %.2f$' % FA[fa_i])
    # plot ground truth lines
    axs[idx].plot([0, 1], [0, 1], 'b', label = 'Simulated f-value')
    axs[idx].legend(loc='upper left')
    axs[idx].set_ylim([-0.1, 1.2])
    axs[idx].set_xlim([-0.1, 1.2])
    axs[idx].set_xlabel('Simulated f-value')
    axs[idx].set_ylabel('Estimated f-value')
```

Similar to what was reported by Hoy et al. (2014), the free water elimination model shows to be able to produce tissue's FA values stable to the free water contaminations for the larger ground truth FA values and when free water contaminations are smaller than 80% of the total volume fraction. Relative to the WLS preformance, the NLS procedure is approximately 1.8 slower than the initial guess.

# Testing different optional parameters of the NLS procedure

```python
fw_params = fw_params_initial.copy()

t0 = time.time()
# NLLS procedures is used to find final parameters guess
fw_params, S0e = nls_fit_tensor(design_matrix, DWI_simulates, 
                                fw_params=fw_params, S0=S0,
                                Diso=Dwater, cholesky=False,
                                f_transform=True, jac=True,
                                mdreg=2.7e-3)

tf = time.time() - t0
print(tf / tr)

# Compute standard measures of the tissue's diffusion tensor
evals = fw_params[..., :3]
fa = dti.fractional_anisotropy(evals)
f = fw_params[..., 12]

median_fa = np.empty(VF.size)
lower_p = np.empty(VF.size)
upper_p = np.empty(VF.size)

# Plot FA values
colors = {0: 'r', 1: 'magenta', 2: 'black', 3: 'b', 4: 'g'}
plt.figure(figsize=(5, 4))
plt.axis([-0.1, 1.2, -0.1, 1.2])
for fa_i in range(FA.size):
    for vf_i in range(VF.size):
        median_fa[vf_i] = np.median(fa[fa_i, vf_i, :])
        p25, p75 = np.percentile(fa[fa_i, vf_i, :], [25 ,75])
        lower_p[vf_i] = median_fa[vf_i] - p25 
        lower_p[vf_i] = p75 - median_fa[vf_i]
    plt.errorbar(VF/100, median_fa, fmt='.',
                 yerr=[lower_p, lower_p],
                 color=colors[fa_i],
                 ecolor=colors[fa_i],
                 linewidth=1.0,
                 label='$FA: %.2f$' % FA[fa_i])
plt.xlabel('Simulated f-value')
plt.ylabel('Estimated FA')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

# Plot F values
median_f = np.empty(VF.size)
lower_p = np.empty(VF.size)
upper_p = np.empty(VF.size)

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

for idx, fa_i in enumerate([0, 4]):
    for vf_i in range(VF.size):
        median_f[vf_i] = np.median(f[fa_i, vf_i, :])
        p25, p75 = np.percentile(f[fa_i, vf_i, :], [25 ,75])
        lower_p[vf_i] = median_f[vf_i] - p25 
        lower_p[vf_i] = p75 - median_f[vf_i]
    axs[idx].errorbar(VF/100, median_f, fmt='.',
                      yerr=[lower_p, lower_p],
                      color=colors[fa_i],
                      ecolor=colors[fa_i],
                      linewidth=3.0,
                      label='$FA: %.2f$' % FA[fa_i])
    # plot ground truth lines
    axs[idx].plot([0, 1], [0, 1], 'b', label = 'Simulated f-value')
    axs[idx].legend(loc='upper left')
    axs[idx].set_ylim([-0.1, 1.2])
    axs[idx].set_xlim([-0.1, 1.2])
    axs[idx].set_xlabel('Simulated f-value')
    axs[idx].set_ylabel('Estimated f-value')
```

Indeed, adding the jacobian analytical solution seem to increase the speed performace of the NLS procedures (now computing time is approximately 1.4 slower than the WLS solution). Nevertheless, it is important to note that results are similar to its initial guess. To insure that the NLS finds the global minimal instead of converging to the local minima position nearest to the initial guess, we decided to keep the use of the Jacobian function as optional.
